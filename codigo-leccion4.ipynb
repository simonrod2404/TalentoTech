```python
#Generar datos aleatorios con demostración
import numpy as np
np.random.seed(0)
X = np.random.rand(100, 2)

print('Datos generados aleatoriamente')
print(X[:5])
```

```python
from scipy.spatial.distance import euclidean

#Calcular la distancia euclidiana entre dos puntos

point1 = [1, 2]
point2 = [4, 6]
distance = euclidean(point1, point2)
print('Distancia euclidiana entre los puntos 1 y 2:', distance)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Inicializar y ajustar el modelo K-Means

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
#Visualizar los clusters

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', s=200)
plt.title('Clustering con K-Means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Aplicación de CLustering en un Conjunto de Datos Real+
import sklearn
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#Cargar conjunto de datos Iris
iris = load_iris()

X = iris.data
y = iris.target

#Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Reducir la dimensionalidad con PCA para visualización
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

#Inicializar y ajustar el modelo K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_scaled)

#Visualizar los clusters
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('Clustering con K-Means en el conjunto de datos Iris')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.show()
```
