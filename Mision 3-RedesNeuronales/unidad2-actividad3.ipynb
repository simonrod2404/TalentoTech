```python
#Importando librerias
import keras
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

#Generando dataset de clasificacin sintetico
X, y = make_classification(n_samples=100, n_features=5, n_classes=2)

#Dividiendo datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#Codificando las etiquetas con un one-hot encoder
y_train_categorical = keras.utils.to_categorical(y_train)
y_test_categorical = keras.utils.to_categorical(y_test)
```

```python
#Creando una instancia de un modelo secuencial de Keras
model = keras.models.Sequential()

#Aadiendo una capa densa con 10 neuronas y funcin de activacin relu
model.add(keras.layers.Dense(5, input_shape=(5,), activation='relu'))

#Aadiendo una capa densa con 10 neuronas y funcin de activacin relu
model.add(keras.layers.Dense(5, activation='relu'))

#Aadiendo una capa de salida con 2 neuronas (una para cada clase) y activacin softmax
model.add(keras.layers.Dense(2, activation='softmax'))

#Compilando el modelo con optimizador adam y funcin de prdida categorical_crossentropy
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

#Entrenando el modelo
model.fit(X_train, y_train_categorical, epochs=300, batch_size=10)
```

```python
#Evaluando el modelo en el conjunto de prueba

loss, accuracy = model.evaluate(X_test, y_test_categorical) 
print('Perdida:', loss)
print('Precisin:', accuracy)
```
